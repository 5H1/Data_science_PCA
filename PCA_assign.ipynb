```python
#==================================================================
# K - means clustering
#==================================================================

import pandas as pd
df = pd.read_csv("wine.csv")
df

from sklearn.preprocessing import StandardScaler
SS = StandardScaler()
SS_X = SS.fit_transform(X)

from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2,n_init=20)

df["kmeans"] = kmeans.fit_predict(SS_X)
df["kmeans"].value_counts()

```

```python

#==============================================================
from sklearn.metrics import silhouette_score

scores = []

for i in range(2,10):
    cluster = KMeans(n_clusters=i,n_init=20)
    df["kmeans"] = cluster.fit_predict(SS_X)
    Y = df["kmeans"]
    scores.append(silhouette_score(SS_X,Y))
    
```

```python
len(scores)
```

```python
#printing silhouette scores for individual clusters
```

```python
print("order wise silhouette score from k=2 to k=9")

for i in range(0,8):
    print(scores[i])
```

```python
import matplotlib.pyplot as plt
plt.scatter(range(2,10),scores)
plt.plot(range(2,10),scores,color='red')
plt.xlabel("cluster numbers")   
plt.ylabel("silhoutte scores")
plt.show()
```
